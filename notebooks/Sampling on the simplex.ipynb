# Sampling (uniformly) on the simplex

Suppose you want to sample positive triplets that sum to one. A common initial approach is to uniformly sample three real numbers from the interval [0, 1], and then normalize the resulting triplet by dividing each number by their total sum. However, this method does not guarantee that the sampled triplets will be uniformly distributed in the space of possible triplets. In other words, it does not account for the specific topology of the space of triplets with a fixed sum of 1.

To achieve a uniform distribution in this space, one can sample from exponential distributions and then normalize, achieving a uniform distribution on the simplex.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot(x):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x[:, 0], x[:, 1], x[:, 2], c='C1', marker='+', lw=0.8)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.view_init(30, 45, 0)
    plt.show()
```

```python
N = 2500
```

First strategy: sampling uniformly over [0, 1] and then normalizing.

```python
x = np.random.rand(N, 3)
x /= x.sum(axis=1)[:, None]

plot(x)
```

Second strategy: sampling from an exponential law and then normalizing.

```python
xt = np.random.exponential(size=(N, 3))
xt /= xt.sum(axis=1)[:, None]

plot(xt)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta
```

```python
from scipy.stats import beta

# Paramètres de la distribution bêta
alpha_param = 5
beta_param = 5

# Générer des valeurs à partir de la distribution bêta
samples = beta.rvs(alpha_param, beta_param, size=10000)

# Calculer la fonction de densité de probabilité (PDF) pour différentes valeurs de p1
x = np.linspace(0, 1, 100)
pdf = beta.pdf(x, alpha_param, beta_param)

# Afficher l'histogramme des échantillons et la fonction de densité de probabilité
plt.hist(samples, bins=30, density=True, alpha=0.5, color='b', label='Échantillons')
plt.plot(x, pdf, 'r-', label='PDF')
plt.title('Distribution bêta')
plt.xlabel('p1')
plt.ylabel('Probabilité')
plt.legend()
plt.show()
```
